## Data Cleaning

```python
cars.make.unique()
```

We will convert the 'make' column into company name column by taking only the firts words of each car.

```python
Company = cars['make'].apply(lambda x: x.split(' ')[0])
cars.insert(3,'Company',Company)
cars.drop(['make'],axis = 1,inplace = True)
```

```python
cars.head()
```

```python
cars.Company.unique()
```

We can see there are some spelling mistakes like:
    maxda = mazda
    Nissan = nissan
    porsche = porcshce
    toyota = toyouta
    vokswagan = volkswagan = vw
So we have to correct these mistakes

```python
cars.Company = cars.Company.str.lower()

cars.Company.replace('maxda','mazda',inplace = True)
cars.Company.replace('toyouta','toyota',inplace = True)
cars.Company.replace('vokswagen','volkswagen',inplace = True)
cars.Company.replace('vw','volkswagen',inplace = True)
cars.Company.replace('porcshce','porsche',inplace = True)
```

### Transforming Categorical Data

```python
def ohe(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df


new_cars = ohe('fueltype',new_cars)
new_cars = ohe('aspiration',new_cars)
new_cars = ohe('carbody',new_cars)
new_cars = ohe('drivewheel',new_cars)
new_cars = ohe('enginetype',new_cars)
new_cars = ohe('cylindernumber',new_cars)
```

# MODEL 1

### Train Test Split and Data Scaling

```python
from sklearn.model_selection import train_test_split
```

```python
x = new_cars.iloc[ : , 1: ]
y = new_cars.price
```

### Data Scaling

```python
from sklearn.preprocessing import scale
x = scale(x)
```

```python
x
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 100)
```

### Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

### Data Training and Testing

```python
model.fit(X_train,y_train)
```

```python
predict = model.predict(X_test)
predict
```

### Model Evaluation

```python
from sklearn.metrics import r2_score, accuracy_score
```

```python
r2_score(y_test,predict)
```

# MODEL 3

## RFE with Cross Validation

```python
from sklearn.feature_selection import RFECV
```

```python
x = new_cars.iloc[ : , 1: ]
y = new_cars.price
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 47)
```

```python
rfecv = RFECV(estimator = model, step = 1, cv = 5)
rfecv = rfecv.fit(X_train, y_train)

print('Qptimal Number Of Features: ', rfecv.n_features_)
print('Best Features: ', X_train.columns[rfecv.support_])
```

```python
rfecv.grid_scores_
```

```python
plt.figure(figsize = (15,6))
plt.xlabel('Number Of features Selected')
plt.ylabel('Crossvalidation Score Of NumberOf Selected Features')
plt.plot(range(1,len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

```python
X_train_rfecv = rfecv.transform(X_train)
X_test_rfecv = rfecv.transform(X_test)
```

### Training and Teating the Model

```python
rfecv_model = model.fit(X_train_rfecv,y_train)
```

```python
rfecv_predict = rfecv_model.predict(X_test_rfecv)
```

### Evaluating The Model

```python
r2_score(y_test,rfecv_predict)
```
