```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cars = pd.read_csv('AutoData.csv')
cars.head()
```

```python
cars.info()
```

```python
cars.iloc[:,9:16]
```

```python
Company = cars['make'].apply(lambda x: x.split(' ')[0])
cars.insert(3,'Company',Company)
cars.drop(['make'],axis = 1,inplace = True)
```

```python
cars.Company = cars.Company.str.lower()

cars.Company.replace('maxda','mazda',inplace = True)
cars.Company.replace('toyouta','toyota',inplace = True)
cars.Company.replace('vokswagen','volkswagen',inplace = True)
cars.Company.replace('vw','volkswagen',inplace = True)
cars.Company.replace('porcshce','porsche',inplace = True)
```

```python
cars.head()
```

```python
cars['Company'].unique()
```

```python
temp = pd.get_dummies(cars['Company'],drop_first  = True)
temp.shape
```

```python
cars.info()
```

```python
new_cars = cars
```

```python
def ohe(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df

new_cars = ohe('symboling',new_cars)
new_cars = ohe('fueltype',new_cars)
new_cars = ohe('Company',new_cars)
new_cars = ohe('aspiration',new_cars)
new_cars = ohe('doornumber',new_cars)
new_cars = ohe('carbody',new_cars)
new_cars = ohe('drivewheel',new_cars)
new_cars = ohe('enginelocation',new_cars)
new_cars = ohe('enginetype',new_cars)
new_cars = ohe('cylindernumber',new_cars)
new_cars = ohe('fuelsystem',new_cars)
```

```python
new_cars['Price'] = new_cars.price
```

```python
new_cars.drop('price', axis = 1, inplace = True)
```

```python
new_cars.columns
```

```python
new_cars.shape
```

# SelectKBest

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
X = new_cars.iloc[:,:68]
y = new_cars.iloc[:,-1]

# Always make sure that the x & y variables have same dtypes so that they there is no value error

X = X.astype(int)
y = y.astype(int)
```

```python
bestfeatures = SelectKBest(score_func = chi2, k = 10) 
fit = bestfeatures.fit(X,y)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
```

```python
featureScores
```

```python
featureScores['Specs'].unique()
```

```python
# Selecting top features with highest scores
# Remember higher the score better the feature
BestFea = featureScores.nlargest(60,'Score')
BestFea
```

```python
BestFea['Specs'].unique()
```

# Scaling the data

```python
xdep = df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
xdep = scaler.fit_transform(xdep)
```

## Spliting into training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(xdep,y, test_size = 0.2, random_state = 100)
```

# Linear Regression model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
predict = model.predict(X_test)
predict
```

## Model Evaluation

```python
from sklearn.metrics import r2_score, accuracy_score
r2_score(y_test,predict)
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train,y_train)
```

```python
rfrpredict = rfr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
r2_score(y_test,rfrpredict)
```

```python
param_grid = {'n_estimators' : [40,80,100,130],
              'max_features':['auto','sqrt','log2'],
              'max_depth':[20,30,40,None],
              'min_samples_split':[1,2,5,7]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_search = RandomizedSearchCV(rfr, param_distributions = param_grid, n_iter = 10, cv = 10)
```

```python
random_search.fit(X_train, y_train) 
```

```python
print('Best score for data:', random_search.best_score_)
print('Best Criterion:', random_search.best_estimator_.n_estimators)
print('Best Max Depth:', random_search.best_estimator_.max_depth)
print('Best Max Features:', random_search.best_estimator_.max_features)
print('Best Min samples split:', random_search.best_estimator_.min_samples_split)
```
